```python
import numpy as np
import pandas as pd
from scipy import optimize #bibliotecade otimização do python. olhar: https://pablormier.github.io/2017/09/05/a-tutorial-on-differential-evolution-with-python/
```

```python
'''Rosembrock Function'''
def function(x):
    fun=0
    a=1.0
    b=100.0
    fun = (a-x[0])**2 + b*(x[1]-x[0]**2)**2
    return fun
```

```python
x0 = np.array([2., 2.]) #ponto aleatório para iniciar
# Initial guess.
```

```python
bounds = [(-5, 5), (-5, 5)] #limtes fronteiras, busca valores de x1 e x2 entre -5 e 5
ret = optimize.differential_evolution(function,bounds,
                                      init="latinhypercube", # obs: init="random" #é uma forma de distribuir os numeros uniformemente num quadrado
                                      maxiter=200, popsize=40,recombination=0.5) #200 iteracoes, um quadradode 40 pontos e  
```

```python
print("global minimum: x = ",ret.x, "f(x0) = ",ret.fun)
```
