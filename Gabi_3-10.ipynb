```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# funcao normatiza dados
def Normatiza(x):
    strings=['SL','time','TC'] #vou utilizar a normatização somente para essas 3 colunas. variáveis independentes
    for i in strings:
        max_x=x[i].max()
        min_x=x[i].min()
        x_M=(max_x+min_x)/2
        x[i]=(x[i]-x_M)/(max_x-x_M)
    return x
```
